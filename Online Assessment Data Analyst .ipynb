# Question 2 Does the day of the week and time of posting affect engagemnt 

```python
Jan= pd.read_csv("C:/Users/PC/Data Analyst Assesment/Jan.csv")
Jan= Jan.loc[Jan['Total Impressions']!=0].copy() #Gets rid of rows that have 0 Total Impressions
Jan['Published Date'] = pd.to_datetime(Jan['Published Date'], format ='%m/%d/%Y %H:%M')
Jan['Day']=Jan['Published Date'].dt.day
Jan['hour']=Jan['Published Date'].dt.hour
Jan['Ratio']= Jan['Total Engagements']/Jan['Total Impressions']#percent of engagements over impressions
```

# Question #3 How are our game titles doing in terms of social performance?
## Answer: It dependes on the platform and media type.

```python
new_df = df.loc[df['Total Impressions']!=0].copy() #Gets rid of rows that have 0 Total Impressions
new_df['Published Date'] = pd.to_datetime(new_df['Published Date'], format ='%m-%d-%Y %H:%M')
# Converts Published Date to a datetime series
Ratio = new_df['Total Engagements']/new_df['Total Impressions']
new_df['Ratio']= Ratio
new_df['day']=new_df['Published Date'].dt.day
new_df['hour']=new_df['Published Date'].dt.hour
new_df.groupby('day').max().sort_values(by='Total Impressions', ascending = False).copy()
new_df.groupby('hour').max().sort_values(by = 'Total Impressions', ascending= False)
new_df.head(5)
```

```python
CSGO=new_df[new_df['Account']=='CSGO']
DOTA2=new_df[new_df['Account']=='DOTA2']
Valorant=new_df[new_df['Account']=='Valorant']
```

## CSGO only had Twitter and one campaign, being Community Engagement. There were also several N/A posts. Most of the posts on were photos, then videos, and texts. There were only two links posted. However, videos had a higher rate of engagement than photos.

```python
CSGO.hist(column='Ratio', by = 'Campaign Name', bins=50)
CSGO.hist(column='Ratio', by = 'Media Type', bins=10)
print('CSGO Photo Count ',CSGO['Media Type'].value_counts()['Photo'])
print('CSGO Text Count ',CSGO['Media Type'].value_counts()['Text'])
print('CSGO Video Count',CSGO['Media Type'].value_counts()['Video'])
print('CSGO Link Count',CSGO['Media Type'].value_counts()['Link'])
```

## Valorant's only platform is Youtube, and had only one campaign. Evil Exhibited had a hard time passing .10% engagement. There were also two N/A post. Expanding to other platforms such as TikTok would greatly benefit Valorant.

```python
Valorant.hist(column='Ratio', by = 'Campaign Name', bins=20)
Valorant.hist(column='Ratio', by = 'Media Type', bins=50)
print('Valorant Video Count ',CSGO['Media Type'].value_counts()['Video'])
```

# Dota 2 had  mainly Twitter and some Youtube videos. The two campaigns on Twitter where  Community Engagement and Evil Exhibited. Most of the post were N/A

```python
DOTA2.hist(column='Ratio', by = 'Campaign Name', bins=50)
DOTA2.hist(column='Ratio', by = 'Media Type', bins=50)
print('Dota Photo Count ',DOTA2['Media Type'].value_counts()['Photo'])
print('Dota Text Count ',DOTA2['Media Type'].value_counts()['Text'])
print('Dota Video Count ',DOTA2['Media Type'].value_counts()['Video'])
print('Dota Link Count ',DOTA2['Media Type'].value_counts()['Link'])
print('Dota Mixed Count ',DOTA2['Media Type'].value_counts()['Mixed'])
```
